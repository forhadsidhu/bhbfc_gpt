```python
import pdfplumber

def pdf2text(pdf_path):
  # Open the PDF from Google Drive
  with pdfplumber.open(pdf_path) as pdf:
      full_text = ""
      for page in pdf.pages:
          full_text += page.extract_text()

  # Print extracted text
  print(full_text)
  return full_text
```

```python
import re
import json

def extract_qa_pairs(text):
    # Define a pattern to match the questions (questions end with '?')
    question_pattern = r'([A-Za-z0-9\s\.,;\'\"\(\)-]+[?])'

    # Define a pattern to capture the answer as everything that follows a question
    # until the next question is encountered
    answer_pattern = r'([^\?]+)(?=\n[A-Za-z])'

    # Find all questions using the question pattern
    questions = re.findall(question_pattern, text)

    # Find all answers using the answer pattern
    answers = re.findall(answer_pattern, text, flags=re.DOTALL)

    # Combine the questions and answers in pairs
    qa_pairs = list(zip(questions, answers))

    return qa_pairs



# Extracting question-answer pairs
qa_pairs = extract_qa_pairs(full_text)

# Convert the QA pairs into Mistral JSON-like format (adjust based on specific Mistral format)
mistral_data = []
for question, answer in qa_pairs:
    mistral_data.append({
        "question": question.strip(),
        "answer": answer.strip()
    })

# Save the QA pairs to a JSON file (Mistral format)
mistral_filename = '/content/drive/My Drive/BHBFC_chatbot/dataset/qa_pairs_mistral.json'
with open(mistral_filename, 'w') as json_file:
    json.dump(mistral_data, json_file, indent=4)

print(f"QA pairs saved to {mistral_filename}")
```

```python
import fitz  # PyMuPDF

def pdf2text_pymupdf(pdf_path):
    # Open the PDF using PyMuPDF
    with fitz.open(pdf_path) as doc:
        full_text = ""
        for page in doc:
            full_text += page.get_text("text")  # Extract text as plain text

    # Print extracted text
    print(full_text)
    return full_text
```

```python
from pdf2image import convert_from_path
import pytesseract

# Path to Tesseract (the installation path in Colab)
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

def extract_text_with_ocr(pdf_path):
    # Convert PDF pages to images
    pages = convert_from_path(pdf_path)

    text = ""

    # Perform OCR on each page
    for page in pages:
        text += pytesseract.image_to_string(page, lang='ben')  # 'ben' is the language code for Bengali

    # Print the extracted text
    print(text)
    return text

# Path to your PDF (you may upload it or use Google Drive)
pdf_path = "/content/drive/My Drive/BHBFC_chatbot/dataset/bangla_one.pdf"

# Extract text using OCR
full_text = extract_text_with_ocr(pdf_path)
```
